### Extra: ¿Qué es el CI y qué verifican estas herramientas?

**CI (Integración Continua)** es una práctica en la que cada vez que se sube código al repositorio (mediante un push o un Pull Request), se ejecutan automáticamente una serie de verificaciones para asegurar que el código cumple ciertos estándares de calidad. En este curso, el CI está configurado en `.github/workflows/ci.yml` y es ejecutado por GitHub Actions.

Las herramientas que configuramos cumplen roles complementarios:

| Herramienta | ¿Qué verifica? |
|---|---|
| **`ruff` (linter)** | Detecta errores de estilo, imports no usados, variables indefinidas, patrones propensos a bugs y sintaxis desactualizada |
| **`ruff` (formatter)** | Formatea el código automáticamente para que tenga un estilo consistente (espacios, saltos de línea, comillas, etc.) |
| **`pre-commit`** | Ejecuta un conjunto de revisiones automáticamente *antes de cada commit*, impidiendo subir código que no pase las verificaciones |

En la práctica: cuando hagan `git commit`, `pre-commit` correrá `ruff` sobre sus archivos. Si hay problemas, el commit será bloqueado y deberán corregirlos antes de continuar. Esto asegura que todo el código que llegue al repositorio sea consistente y libre de errores básicos.

#### Pregunta 1 (0.5 pts)

Dada una lista de $n$ elementos, no necesariamente ordenada, defina funciones que busquen:
- El elemento menor.
- El elemento mayor.
- El promedio de los elementos en la lista.

Ejemplo:
```python
lista_n = [12,4,34,1,2,52,1,54,98,5,235,76,57,87,123,32,32]
menor = 1
mayor = 235
```


````python
def min_mds7202(lista):
    """
    Calcula el valor mínimo de lista
    :param lista: lista
    :return: valor mínimo
    """
    x = float("inf")
    for valor in lista:
        if valor < x:
            x = valor
    return x


def max_mds7202(lista):
    """
    Calcula el valor máximo de lista
    :param lista: lista
    :return: valor máximo
    """
    x = float("-inf")
    for valor in lista:
        if valor > x:
            x = valor
    return x


def mean_mds7202(lista):
    """
    Calcula el promedio de lista
    :param lista: lista
    :return: valor promedio
    """
    return sum(lista) / len(lista)
````

````python
lista_n = [12, 4, 34, 1, 2, 52, 1, 54, 98, 5, 235, 76, 57, 87, 123, 32, 32]
menor = min_mds7202(lista_n)
mayor = max_mds7202(lista_n)
promedio = mean_mds7202(lista_n)

print(
    f" menor: {menor}\n",
    f"mayor: {mayor}\n",
    f"promedio: {promedio}",
)
````

#### Pregunta 2 (0.5 pts)
Genere una función que obtenga la moda de una lista usando diccionarios.

````python
def moda(lista):
    """
    Obtiene la moda de la lista entregada usando diccionarios.
    :param lista: lista
    :return: valor mediana
    """
    diccionario = {}
    for valor in lista:
        if valor not in diccionario:
            diccionario[valor] = 1
        else:
            diccionario[valor] += 1

    mayor_cuenta = 0
    x = None
    for valor in diccionario:
        cuenta = diccionario[valor]
        if cuenta > mayor_cuenta:
            mayor_cuenta = cuenta
            x = valor

    return x
````

````python
lista_m = [0, 0, 0, -1, -1, 2, 2, 2, 2, 1]
print("moda:", moda(lista_m))
````

#### Pregunta 3 (0.5 pts)
Vea si un número es palíndromo o no. Los números palíndromos son aquellos que al leerlos al revés y al derecho son iguales, por ejemplo el 123321 o el 98789.

````python
def palindromo(x):
    """
    Verifica si un número es palíndromo
    :param x: número entero
    :return: booleano que indica si es palíndromo
    """
    x_str = str(x)
    return x_str == x_str[::-1]
````

````python
print(
    f" ¿Es palíndromo 123321? {palindromo(123321)}\n",
    f"¿Es palíndromo 98789? {palindromo(98789)}\n",
    f"¿Es palíndromo 1234? {palindromo(1234)}",
)
````

#### Pregunta 4 (0.5 pts)
Dado un texto, conviértalo en estilo Camel Case, vea el siguiente ejemplo:

    Texto inicial: “Este es un texto de ejemplo para mostrar el Camel Case”.

    Texto final: “EsteEsUnTextoDeEjemploParaMostrarElCamelCase”.

````python
def camel_case(texto):
    """
    Convierte un texto en estilo Camel Case
    :param texto: cadena de texto
    :return: texto en estilo Camel Case
    """
    texto_split = texto.split(" ")
    x = ""
    for palabra in texto_split:
        x += palabra.capitalize()
    return x
````

````python
texto_inicial = "Este es un texto de ejemplo para mostrar el Camel Case"
print(f' Texto inicial: “{texto_inicial}”.\\n", f"Texto final: “{camel_case(texto_inicial)}”.')
````

#### Pregunta 1 (0.5 pts):

Un monto inicial de dinero (por ejemplo un millón de pesos) se puede depositar en un banco con un interés anual de 5%. Esto significa  que  cumplido  un  año  se  dispondrá  de  \\$1.050.000. Si se deposita por dos años, entonces  al  finalizar  el  2º  año se dispondrá de \\$1.102.500,  es  decir,  \\$1.050.000  reajustado  en  5%.  

Al  respecto, si  se  invoca  a  la  función  `calc_capital_final(1000000, 5, 2)`  entrega  el resultado `1050000.0`. 

Escriba la función `calc_capital_final`

$$C_f = C_i(1+i)^n$$

- $C_f$ = Capital Final
- $C_i$ = Capital Inicial
- $i$ = Tasa de interés
- $n$ = Periodo del ahorro

Además, incluya tests que certifiquen el correcto funcionamiento de la función.

````python
def calc_capital_final(capital_inicial, tasa_interes, anos):
    """
    Calcula el capital final basado en la fórmula de interés compuesto.

    Args:
        capital_inicial (float): Ci
        tasa_interes (float): i (en porcentaje, ej: 5 para 5%)
        anos (int): n
    """
    # Convertimos el porcentaje a decimal (5% -> 0.05)
    i_decimal = tasa_interes / 100

    # Aplicamos la fórmula: Cf = Ci * (1 + i)^n
    capital_final = capital_inicial * (1 + i_decimal) ** anos

    return capital_final


# --- Pruebas Unitarias ---
def test_calc_capital_final():
    # Caso 1: 1 año al 5%
    assert calc_capital_final(1000000, 5, 1) == 1050000.0

    # Caso 2: 2 años al 5% (Interés compuesto)
    assert calc_capital_final(1000000, 5, 2) == 1102500.0

    # Caso 3: Sin intereses (0%)
    assert calc_capital_final(500000, 0, 10) == 500000.0

    # Caso 4: Capital inicial cero
    assert calc_capital_final(0, 10, 5) == 0.0

    print("Tests pasaron exitosamente.")


test_calc_capital_final()
````

#### Pregunta 2 (0.5 pts):

Un anagrama es una palabra que resulta de la transposición de letras de otra palabra o frase.

Por ejemplo, las palabras “caso” y “caos” sonan agramas de “saco”  y las palabras “acaso” y “ocaso” no son anagramas de “caso”. Las anagramas poseen las mismas letras, con la misma cantidad de apariciones, pero en un orden diferente

Escriba la función `def anagrama(x,y)` que reciba dos palabras y entregue `True` si son anagramas y `False` si no.


````python
def anagrama(palabra1, palabra2):
    """
    Verifica si dos palabras son anagramas.

    Args:
        palabra1 (str): Primera palabra
        palabra2 (str): Segunda palabra

    Returns:
        bool: True si son anagramas, False en caso contrario
    """
    # Eliminamos espacios y convertimos a minúsculas
    p1 = palabra1.replace(" ", "").lower()
    p2 = palabra2.replace(" ", "").lower()

    # Comparamos las letras ordenadas de ambas palabras
    return sorted(p1) == sorted(p2)


# --- Pruebas Unitarias ---
def test_anagrama():
    # Caso 1: "listen" y "silent" son anagramas
    assert anagrama("listen", "silent")

    # Caso 2: "triangle" y "integral" son anagramas
    assert anagrama("triangle", "integral")

    # Caso 3: "apple" y "pabble" no son anagramas
    assert not anagrama("apple", "pabble")

    # Caso 4: Anagramas con espacios y mayúsculas
    assert anagrama("Dormitory", "Dirty Room")

    print("Tests de anagrama pasaron exitosamente.")


test_anagrama()
````
